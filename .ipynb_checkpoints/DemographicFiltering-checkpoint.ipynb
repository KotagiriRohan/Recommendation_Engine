```python
import pandas as pd
import numpy as np
moviecredits_df = pd.read_csv('Dataset/tmdb_5000_credits.csv')
movie_df  = pd.read_csv('Dataset/tmdb_5000_movies.csv')
```

```python
movie_df.head()
```

```python
moviecredits_df.head()
```

```python
#lets merge the both the datasets
moviecredits_df.columns = ['id','title','cast','crew']
movie_df = movie_df.merge(moviecredits_df,on=["id","title"])
```

```python
movie_df.head(5)
```

<b>Metric For rating</b>

v is the number of votes for the movie;<br>
m is the minimum votes required to be listed in the chart;<br>
R is the average rating of the movie; And <br>
C is the mean vote across the whole report <br>

```python

C = movie_df['vote_average'].mean()
#Minimun no of votes that have more votes than 90% of the movies 
m = movie_df['vote_count'].quantile(0.9)
```

```python
q_movies = movie_df.copy().loc[movie_df['vote_count'] >= m]
q_movies.shape
```

```python
def weighted_rating(x,m=m,C=C):
    v= x['vote_count']
    R = x['vote_average']
    return (v / (v+m)*R) + (m/(m+v)*C)
#applying the above function to get the rating evaluation
q_movies['score'] = q_movies.apply(weighted_rating, axis=1)
#sort the movies according to the weighted rating values 
q_movies = q_movies.sort_values('score',ascending=False)
#Displaying the top 10 movies according to the weighted rating
q_movies[['title','score']].head(10)
```
